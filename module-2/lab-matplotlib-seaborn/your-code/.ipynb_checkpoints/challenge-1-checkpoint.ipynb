```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# https://matplotlib.org/3.1.0/gallery/color/named_colors.html
plt.figure(figsize=(15,5)) # es para cambiar el ancho y alto de los plots en pulgadas
plt.subplot(1,2,1, title = 'Plot (x,y)', fc = 'white')  # subplot(nrow,ncol,position)
plt.plot(x,y, color = 'indigo', lw= 10)
plt.subplot(1,2,2, title = 'Plot (x,z)', fc = 'white')
plt.plot(x,z, color = 'tab:blue', lw = 10);  #lw es para aumentar el tamaño de la linea 
```

```python
fig,axes= plt.subplots(1,2)
axes[0].plot(x,y, color = 'indigo', lw = 5)
axes[1].plot(x,z, color = 'tab:blue', lw = 5);
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
fig,axes= plt.subplots(nrows = 1,ncols = 2, figsize = (15,5))
axes[0].plot(x,y, color = 'indigo', lw = 10)
axes[1].plot(x,z, color = 'tab:blue', lw = 10);
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
fig,axes= plt.subplots(nrows = 1,ncols = 2, figsize = (15,5))
axes[0].plot(x,y, color = 'indigo', lw = 10)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Plot (x,y)')
axes[1].plot(x,z, color = 'tab:blue', lw = 10)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('Plot (x,z)');
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
fig,axes= plt.subplots(nrows = 1,ncols = 2, figsize = (20,5))
axes[0].plot(x,x**2, color = 'indigo')
axes[0].set_xlabel('x')
axes[0].set_ylabel('x^2')
axes[0].set_title('Plot (x,x^2)')
axes[1].plot(x,np.exp(x), color = 'tab:blue')
axes[1].set_xlabel('x')
axes[1].set_ylabel('exp(x)')
axes[1].set_title('Plot (x,exp(x))');
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2,x, np.exp(x))
axes[1].plot(x, x**2,x, np.exp(x))
axes[1].set_yscale("log")
```
